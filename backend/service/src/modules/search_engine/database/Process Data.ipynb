```python
import json
import re 
```

```python
from collections import Counter
  
def removeElements(lst, k):
    counted = Counter(lst)
      
    temp_lst = []
    for el in counted:
        if counted[el] < k:
            temp_lst.append(el)
              
    res_lst = []
    for el in lst:
        if el not in temp_lst:
            res_lst.append(el)
              
    return(res_lst)
```

```python
with open(file_path, "r") as f:
    json_file = f.read()
    json_dict = json.loads(json_file)
```

```python
job_title_list = []
degree_list = []
company_list = []
college_list = []
skill_list = []
```

```python
for json_person in json_dict:
    
    #Process skill
    skill_section = json_person['skill']
    if skill_section:
        skill_list += skill_section
        
    #Process education
    education_list = json_person['education']
    for education in education_list:
        college = education['university']
        if college:
            college_list.append(college)
            degree = education['degree']
            if degree:
                degree  = degree.split("\n")[1]
                degree_list.append(degree.strip())
            
    #Process experience
    experience_list = json_person['experience']
    for experience in experience_list:
        company = experience['company']
        if company:
            #Process Company
            
            if not any([e for e in ['đại học', 'công ty', 'school','university','college', 'polytechnic', 'academy', 'institute'] if e in company.lower()]):
                
                for ele in ['Full-time','Part-time','Internship', 'Freelance', 'Self-employed']:
                    company = company.replace(ele,"")
                    
                company = re.sub("\(.*?\)","",company.replace("\n",""))
                
                company = company.replace("Việt Nam", "Vietnam")
                
                if len([e for e in company if e.isalpha()]) > 4:
                    company_list.append(company.strip())
                
            job_title = experience['position_title']
            if 'at' in job_title:
                job_title = job_title.split('at')[0]
                
            for mark in [',', '&', 'and', '/', '\\']:
                if mark in job_title:
                    job_title = job_title.split(mark)
                    break
                    
            for job in job_title:
                if len([e for e in job if e.isalpha()]) > 4:
                    job_title_list.append(job.strip())
                
```
